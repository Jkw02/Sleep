```python
numerical_columns.corr()
```

The table above shows the correlation between each of the variables. The closer the correlation is to 1 or -1 the stronger the correlation is and the closer to correlation is to 0 the weaker the correlation. If the correlation between two variables is 0 then we can say they have no correlation between each other. If the correlation is postive it suggests as one variable increases/ decreases so does the other. A negative correlation suggests as one variable increases the other one will decreases and vise versa. The leading diagnol is one as it is the same variable agaisnt each other. 

A chart that can be used to visualy see the correlation would be a heatmap. 

```python
plt.figure(5,figsize=(15,15))
# #annot =True - add the correlation figure in each box on the graph
# #cmap="YlGnBu" - gives the colour of the graph 
# #linewidth=0.2 - the width of the lines between eah box 
# #square =True - the graph will be plotted as a sqaure 
sns.heatmap(numerical_columns.astype(float).corr(),annot=True, cmap="YlGnBu",linewidth=0.2,square=True)
plt.title("Heatmap of the Correlation Between Each Variable")
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.show()
```

As shown by the colour chart to the right of the graph the lighter the colour the more negative the correlation and the darker the colour the more positive the correlation. Looking at the graph the two variables with the strongest postive correlation is systolic blood pressure and Diastolic blood pressure. This suggests as one increases so does the other. This would make sense as they are both linked to your blood pressure. The weakest correlation is between quality of sleep and daily steps, this is 0.017. This suggests there is no correlation with somebody having more sleep and more/less daily steps. 

Another high correlation is between sleep duration and quality of sleep which is 0.88. As this is a positive correlation it suggests as sleep duration increases so does the quality of sleep. 

Stress level and quality of sleep are strongly negativly correlated, suggesting as one increases the other will decrease. 

```python
#scatter plot of Weakest and Strongest Correlated Variables found from the heatmap above

plt.figure(6,figsize=(10,10))
plt.subplot(2,1,1)
plt.scatter(x = clean_df['Systolic Blood Pressure'],y =clean_df['Diastolic Blood Pressure'] )
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Diastolic Blood Pressure')
plt.title('Systolic Blood Pressure against Diastolic Blood Pressure')

plt.subplot(2,1,2)
plt.scatter(x = clean_df['Daily Steps'],y =clean_df['Quality of Sleep'] )
plt.xlabel('Daily Steps')
plt.ylabel('Quality of Sleep')
plt.title('Daily Steps agaisnt Quality of Sleep')

plt.suptitle("Scatter Plots of the Weakest and Strongest Correlated Variables")
plt.show()
```

From the scatter plot above it is clear the difference between strongly correlated variables and weakly correlated variables. The first scatter plot which shows the correlation betweeen Systolic Blood Pressure against Diastolic Blood Pressure shows as one increase so does the other. However in the second plot od quality of sleep agaisnt daily steps there is no correlation. The plots in the graph are plotted all over the graph and no trend can be found.

```python
#get the correlation of each variable agaisnt sleep duration and sleep quality

sleep_cor = numerical_columns.corr(method="pearson")['Sleep Duration']
sleep_quality_cor = numerical_columns.corr(method="pearson")['Quality of Sleep']
```

```python
col_names = list(numerical_columns.columns)

plt.figure(7,figsize = (20,20))
plt.subplot(2,1,1)
plt.scatter(col_names,sleep_cor)
plt.xlabel('Variable')
plt.ylabel('Correlation')
plt.title('The Correlation of Each Variable Agaisnt Sleep Duration')
plt.subplot(2,1,2)
plt.scatter(col_names,sleep_quality_cor)
plt.xlabel('Variable')
plt.ylabel('Correlation')
plt.title('The Correlation of each variable agaisnt sleep duration')
plt.suptitle("Correlation Between two Sleep Variables")
plt.show
```

The two scatter plots above shows the correlation of each variable against sleep duration and sleep quality. The highest correlation for both variables were each other, for example for sleep duration it was quality of sleep and for quality of sleep it was sleep duration. Another highly correlation variable for both sleep variables was stress. This was highly negativly correlated suggesting as your stress levels increase your sleep qaulity and duration will descrease. 

The weakest correlation for both variables was both blood pressure variables. This suggest your blood pressure doesnt have much effect on your sleep.

```python
# Grouping data by 'Age' and calculating the average sleep duration for each age group
average_sleep_duration = clean_df.groupby('Age')['Sleep Duration'].mean().reset_index()

# Create a dictionary to map age groups to numerical values for plotting
age_mapping = {age: i for i, age in enumerate(average_sleep_duration['Age'])}

# Map age groups to numerical values in a new column 'AgeNumeric'
average_sleep_duration['AgeNumeric'] = average_sleep_duration['Age'].map(age_mapping)

# %%
# Plotting a bar graph for average sleep duration per age group
plt.figure(9,figsize=(10, 6))
plt.bar(average_sleep_duration['AgeNumeric'], average_sleep_duration['Sleep Duration'], color='skyblue')
plt.title('Average Sleep Duration per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Sleep Duration (hours)')
plt.xticks(average_sleep_duration['AgeNumeric'], average_sleep_duration['Age'], rotation=45)
plt.tight_layout()
plt.show()
```

The graph above breaks down the ages of the data agaisnt sleep duration. It shows that the age group with the highest amount of sleep are 53-54 years old whilst the age groups with the least amount of sleep are 34 and 50. 

```python
# Grouping data by 'Occupation' and calculating the average sleep duration for each occupation
avg_sleep_duration = clean_df.groupby('Occupation')['Sleep Duration'].mean().reset_index()

# Sorting the values by average sleep duration in descending order
avg_sleep_duration = avg_sleep_duration.sort_values(by='Sleep Duration', ascending=False)

# Plotting a bar chart for average sleep duration per occupation
plt.figure(figsize=(10, 6))
plt.bar(avg_sleep_duration['Occupation'], avg_sleep_duration['Sleep Duration'], color='skyblue')
plt.title('Average Sleep Duration per Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Sleep Duration (hours)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

This bar graph shows the average sleep duration split by the occuaptions. It shows that the occuaption of an engineer has the highest average sleep whilst the job of a scientist has the least average sleep.
